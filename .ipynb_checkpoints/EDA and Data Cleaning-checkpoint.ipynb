```python
# importing needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset from an Excel file
df = pd.read_excel(r"D:\OmniFy Assesment\DataAnalyst_Assesment_Dataset.xlsx")
df.head()
```

```python
# checking datatypes
df.info()
```

```python
#statistics of numerical data
df.describe()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#checking unique values in each column
df.nunique()
```

```python
#drop not needed columns
df = df.drop(columns='Subscription Type')
df.head()
```

```python
# Check unique values in each columns
df['Service Name'].unique()
```

```python
df['Facility'].unique()
```

```python
df['Class Type'].unique()
```

```python
df['Booking Type'].unique()
```

```python
df['Service Type'].unique()
```

```python
df[df['Facility'].isna()]['Booking Type'].unique()
```

```python
#filling null values accordingly
df['Facility'] = df['Facility'].fillna('Class')
```

```python
#filling null values accordingly
df['Class Type'] = df['Class Type'].fillna('Not a Class')
```

```python
df.info()
```

```python
#checking this works correctly or not
df['Instructor'].fillna(pd.Series(
    np.where(df['Booking Type'] != 'Class', 'No Need', 'Unknown'), index=df.index
)).value_counts()
```

```python
#filling null values of instructor columns
df['Instructor'] = df['Instructor'].fillna(pd.Series(
    np.where(df['Booking Type'] != 'Class', 'No Need', 'Unknown'), index=df.index
))
```

```python
df.info()
```

```python
# filling null values of Theme Column
df['Theme'] = df['Theme'].fillna('No Theme')
```

```python
df.info()
```

```python
#filling null values in Time Slot accordingly
df['Time Slot'] = df['Time Slot'].fillna('Not Specified')
```

```python
df.info()
```

```python
#filling duration column with median value
df['Duration (mins)'] = df['Duration (mins)'].fillna(df['Duration (mins)'].median())
```

```python
df.info()
```

```python
#filling 0 values of price column with median
df['Price'] = df.groupby('Booking Type')['Price'].transform(lambda x: x.replace(0, x.median()))
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# checking outliers in price column
df[['Price']].boxplot()
plt.show()
```

```python
#adding column of Day Name
df['Day Name'] = df['Booking Date'].dt.day_name()
```

```python
df
```

```python
#storing cleaned data file
df.to_excel("Cleaned_DataAnalyst_Assesment_Dataset.xlsx", index=False)
```
